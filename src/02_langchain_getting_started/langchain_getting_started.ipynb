Now we must create a .env file in the project root, with the following variables:

```
LANGCHAIN_API_KEY=...
LANGCHAIN_TRACING_V2=true
LANGCHAIN_ENDPOINT='https://api.smith.langchain.com'
LANGCHAIN_PROJECT='ollama-langchain-course'
OLLAMA_BASE_URL='http://localhost:11434'
```

The API key must be acquired in https://smith.langchain.com after sign in.

The Langchain project name will be displayed in Langsmith logs.

````python
# Now we need to load the env
from dotenv import load_dotenv

load_dotenv('../../.env')

# Testing the env
import os

assert os.getenv('LANGCHAIN_API_KEY') is not None
assert os.getenv('LANGCHAIN_TRACING_V2') is not None
assert os.getenv('LANGCHAIN_ENDPOINT') is not None
assert os.getenv('LANGCHAIN_PROJECT') is not None
assert os.getenv('OLLAMA_BASE_URL') is not None
assert os.getenv('OLLAMA_LLM_MODEL') is not None


os.getenv('LANGCHAIN_PROJECT')
````

Now we are going to instantiate a Ollama chat model within Lang Chain

````python
from langchain_ollama import ChatOllama
from langchain_core.messages import AIMessage

base_url = os.getenv('OLLAMA_BASE_URL')
model = os.getenv('OLLAMA_LLM_MODEL')

llm = ChatOllama(
    base_url=base_url,
    model=model,
    temperature=0.8,
    num_predict=2560, # number of tokens it is going to predict
)

# Starting a converstion, saving the response
response: AIMessage = llm.invoke('hi!')

print(response.content)
````

````python
# To stream the response, like Chat GPT and others, we may do the following:
for chunk in llm.stream('In few words, how to bake a carrot cake?'):
    print(chunk.content)
````
