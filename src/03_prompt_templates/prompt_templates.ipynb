```python
# Now we need to load the env
from dotenv import load_dotenv

load_dotenv('../../.env')
```

```python
import os
from langchain_ollama import ChatOllama
from langchain_core.messages import AIMessage
base_url = os.getenv('OLLAMA_BASE_URL')
model = 'minimax-m2.5:cloud'

llm = ChatOllama(base_url=base_url, model=model)

question =  'tell me about the earth in 3 points'
response: AIMessage = llm.invoke(question)
print(response.content)
```

## Customizing System Messages
When you define a system message, you will be able to create a default behaviour, making the LLM assume a particular role, that will be used in the subsquent messages sent.

```python
from langchain_core.messages import HumanMessage, SystemMessage

llm = ChatOllama(base_url=base_url, model=model)

system_message = SystemMessage('you are an elementary teacher. You answer in short sentences')
question = HumanMessage('tell me about the earth in 3 points')
response: AIMessage = llm.invoke([system_message, question])

print(response.content)
```

```python
from langchain_core.messages import HumanMessage, SystemMessage

llm = ChatOllama(base_url=base_url, model=model)

system_message = SystemMessage('you are a phd teacher. You answer in short sentences')
question = HumanMessage('tell me about the earth in 3 points')
response: AIMessage = llm.invoke([system_message, question])

print(response.content)
```

## Prompt Templates
We may create a template for human and system messages, so we can alter their behaviour without having to redefine all the prompt again.

For example, say we want to change the type of school teacher or the topic, in runtime.

### Using SystemMessagePromptTemplate

```python
from langchain_core.prompts import SystemMessagePromptTemplate
system_template = SystemMessagePromptTemplate.from_template('You are {school} teacher. You answer in short sentences.')
system_template
```

### Using HumanMessagePromptTemplate

```python
from langchain_core.prompts import HumanMessagePromptTemplate
human_template = HumanMessagePromptTemplate.from_template('Tell me about the {topic} in {points} points.')
human_template
```

### Invoking the LLM with the Human and System templates

```python
system_msg: SystemMessage = system_template.format(school='phd')
human_msg: HumanMessage = human_template.format(topic='sun', points=5)
messages = [system_msg, human_msg]
messages
```

```python
print(llm.invoke(messages).content)
```

### Using ChatPromptTemplate

We may also combine the templates in a `ChatPromptTemplate`, and format all the templates at once:

```python
from langchain_core.prompts import ChatPromptTemplate
template = ChatPromptTemplate([system_template, human_template])
template_question = template.invoke({'school': 'elementary', 'topic': ' mars', 'points': 3})
template_question
```

```python
print(llm.invoke(template_question).content)
```
